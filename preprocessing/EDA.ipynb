##### Country difference

```python


airlines["Country"] = airlines["Country"].str.strip()
airports["Country"] = airports["Country"].str.strip()

name_map = {
     "Hong Kong SAR of China": "Hong Kong",
     "Lao Peoples Democratic Republic": "Laos",
     "Macao": "Macau",
     "Columbia" : "Colombia",
     "Lao People's Democratic Republic": "Laos",
     "Russian Federation": "Russia",
}   
        
name_series = pd.Series(data=list(name_map.values()), index=list(name_map.keys()))
name_series
```

```python
purge_list = list(name_map.keys())
idx = airports.index[airports["Country"].isin(purge_list)]
a_idx = airlines.index[airlines["Country"].isin(purge_list)]


airports_problems = airports.loc[idx, "Country"]
airlines_problems = airlines.loc[a_idx, "Country"]

airports_fixed = name_series.loc[airports_problems.values].values
airlines_fixed = name_series.loc[airlines_problems.values].values 

airports.loc[idx, "Country"] = airports_fixed
airlines.loc[a_idx, "Country"] = airlines_fixed
```

### remving some entities

```python
airports_countries = airports["Country"].unique().tolist()
airlines_countries = airlines["Country"].unique().tolist()


airlines_countries = set(airlines_countries)
airports_countries = set(airports_countries)

diff = airports_countries.difference(airlines_countries)
a_diff = airlines_countries.difference(airports_countries)

total_diff = list(diff) + list(a_diff)

airports = airports.loc[~airports["Country"].isin(total_diff), :]
airlines = airlines.loc[~airlines["Country"].isin(total_diff), :]
```
